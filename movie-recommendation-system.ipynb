```python
# Reading the 'tmdb_5000_movies.csv' file into a pandas DataFrame named 'movies'
# This file likely contains information about movies, such as titles, genres, budget, revenue, etc.
movies = pd.read_csv('tmdb_5000_movies.csv') 
# Reading the 'tmdb_5000_credits.csv' file into another pandas DataFrame named 'credits'
# This file likely contains information about movie credits, such as cast and crew details
credits = pd.read_csv('tmdb_5000_credits.csv')

```

```python
# The 'merge()' function in pandas is used to combine two DataFrames based on a common column.
# Here, we are merging the 'movies' DataFrame with the 'credits' DataFrame.
# The 'on='title'' specifies that the merge will be performed based on the 'title' column, 
# which is assumed to be present in both DataFrames.
movies = movies.merge(credits,on='title')
```

```python
#genres 
#id
#title
#keywords
#title
#overview
#cast
#crew in crew there is director information some people watch movies on the basis of there favourite director.
# This line is used to select specific columns from the 'movies' DataFrame.
# The syntax 'movies[['col1', 'col2', ...]]' creates a new DataFrame that contains only the specified columns.

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# The 'ast' module in Python stands for "Abstract Syntax Tree".
# It provides tools to work with Python's internal representation of code,
# including parsing, analyzing, and manipulating Python code.
# The module can be useful for tasks like converting strings to Python objects
# or analyzing the structure of Python code.

import ast
```

```python
def convert(text):
    L = []  # Initialize an empty list to store the result
    
    # Convert the input 'text' string to a Python literal (e.g., a list of dictionaries) using ast.literal_eval
    for i in ast.literal_eval(text):  # Iterate over each item in the evaluated list (assumed to be a list of dictionaries)
        # For each dictionary (i), extract the value associated with the 'name' key and append it to the list L
        L.append(i['name'])  
    
    # Return the list L containing all the 'name' values
    return L
```

```python
movies.dropna(inplace=True) 
# Remove all rows with missing values (NaN) from the DataFrame 'movies'
# The 'dropna()' method is used to drop any rows that contain at least one NaN value.
```

```python
movies['genres'] = movies['genres'].apply(convert)
# After applying the 'convert' function to the 'genres' column, we display the first few rows of the updated DataFrame.
# This is useful to check the changes made by the function.
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python

#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python

movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
# The 'apply()' function is used to apply a transformation to each element in the 'overview' column.
# The lambda function inside 'apply()' takes each value (a string) from the 'overview' column and splits it into a list of words.
# The 'split()' method breaks a string into a list based on spaces (by default) between words.
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
# This line combines multiple columns ('overview', 'genres', 'keywords', 'cast', and 'crew') into a single new column 'tags'.
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
# This line applies a transformation to the 'tags' column of the 'new' DataFrame.

new.head()
```

```python
# Importing the CountVectorizer class from sklearn's feature_extraction.text module.
# CountVectorizer is used to convert a collection of text documents into a matrix of token counts.
# It also provides options to filter out stop words and limit the number of features (words).
from sklearn.feature_extraction.text import CountVectorizer
# Creating an instance of CountVectorizer with the following options:
# - 'max_features=5000': Limits the number of features (words) to the top 5000 most frequent words in the dataset.
# - 'stop_words='english'': Removes common English stop words (e.g., 'the', 'is', 'in', etc.) from the text data.
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python


# This line applies the CountVectorizer to the 'tags' column in the 'new' DataFrame.
# The 'fit_transform()' method learns the vocabulary from the 'tags' column and transforms the text into a feature matrix.
# The 'toarray()' method converts the resulting sparse matrix into a dense numpy array.
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
# Importing the cosine_similarity function from sklearn.metrics.pairwise.
# Cosine similarity is a metric used to measure how similar two vectors are, based on the cosine of the angle between them.
# It is commonly used to compare text data, especially when vectors represent text documents, like the movie features in this case.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# This line calculates the cosine similarity between all pairs of movies based on their feature vectors.
# 'vector' contains the feature matrix obtained from the 'tags' column after applying CountVectorizer.
# cosine_similarity() computes the similarity between the feature vectors, returning a similarity matrix.

similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
# This function provides movie recommendations based on the cosine similarity between movies.
# It takes the title of a movie as input, finds the index of that movie in the 'new' DataFrame,
# and then uses the similarity matrix to recommend the top 5 most similar movies.

def recommend(movie):
    # Find the index of the movie in the 'new' DataFrame based on its title.
    index = new[new['title'] == movie].index[0]
    
    # Get the similarity scores of the movie at the found index.
    # The similarity scores are sorted in descending order based on the similarity value.
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    
    # Loop through the sorted list of distances (skipping the first element, which is the movie itself).
    # The top 5 most similar movies are printed.
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python

recommend('Avatar')
```
